```python
from scipy.stats import norm
import numpy as np

n = 250
x = 70
p_hat = x/n

z_norm = norm(loc=0, scale=1)
ll = p_hat - z_norm.ppf(0.975) * np.sqrt(p_hat*(1-p_hat)/n)
ul = p_hat + z_norm.ppf(0.975) * np.sqrt(p_hat*(1-p_hat)/n)

print("The point estimator of the proportion is ", p_hat)
print("95% confidence interval of the proportion is {} and {}".format(np.round(ll,4), np.round(ul,4)))
```

```python
n = 250
x =70
p_hat = x/n

z_norm = norm(loc = 0 , scale =1)
ll = p_hat - z_norm.ppf(0.95)*np.sqrt(p_hat*(1-p_hat)/n)
ul = p_hat + z_norm.ppf(0.95)*np.sqrt(p_hat*(1-p_hat)/n)

print("The point estimator of the proportion is ", p_hat)
print("90% of the confidence interval of the proportion is {} and {}".format(np.round(ll,4), np.round(ul,4)))
```

```python
95%신뢰구간은 90%신뢰구간보다 더 넓다.
더 높은 확실성을 얻기 위해 정밀성이 줄어들었다.
신뢰 수준이 낮아진만큼 추정범위가 정밀해지고,
신뢰 수준이 높아진만큼 추정범위가 조잡해진다.
```
