```python
import math
import scipy.stats as st

#오류크기
alpha = 0.10
#표준편차
sigma = 18.8
#허용최대오차
margin_of_error = 0.27

#임계값계산
z_critical = st.norm.ppf(1 - alpha / 2)

#표본크기 계산공식적용
sample_size_raw = (z_critical * sigma / margin_of_error)**2

#반올림
min_sample_size = math.ceil(sample_size_raw)

print(f"주어진 조건을 만족하는 최소 표본의 크기(n)는 {min_sample_size}입니다.")
```
