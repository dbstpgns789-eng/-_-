```python
import numpy as np
import matplotlib.pyplot as plt

num_flips = 1000
outcomes = [0, 1] #0 뒷면 , 1 앞면

flips = np.random.choice(outcomes, size=num_flips)
cuml_heads = np.cumsum(flips)
prob = cuml_heads / np.arange(1, num_flips + 1)

plt.figure(figsize=(10, 6))
plt.plot(np.arange(1, num_flips + 1), prob, 'b-o', markersize=5, lw=1)
plt.axhline(y=0.5, color='r', linestyle='-', lw=2)
final_prob = prob[-1]
plt.title(f'Probability is {final_prob:.3f}', fontsize=14)
plt.ylim(-0.05, 0.95) 
plt.xlim(-50, 1050)
plt.tick_params(axis='both', which='major', labelsize=12)

plt.show()
```
